```python
player1 = player2 = XorO_P1 = XorO_P2 = None
board = [x for x in "         "]

def choose_player():
    global player1, player2, XorO_P1, XorO_P2
    player1 = input("Enter player 1 name : ")
    XorO_P1 = input("Do you want to be X or O ?")
    player2 = input("Enter player 2 name : ")
    XorO_P2 = 'X' if XorO_P1=='O' else 'O'

def print_board(board):
    print(" {0}|{1}|{2}\n-------\n {3}|{4}|{5}\n-------\n {6}|{7}|{8}"
        .format(board[0], board[1], board[2], board[3], board[4], board[5], board[6], board[7], board[8]))

def win_check(board,mark):
        print(board)
        print(mark)

        if ((board[6] == mark and board[7] == mark and board[8] == mark) or (board[3] == mark and board[4] == mark and board[5] == mark) or (board[0] == mark and board[1] == mark and board[2] == mark) or (board[0] == mark and board[3] == mark and board[6] == mark) or (board[1] == mark and board[4] == mark and board[7] == mark) or (board[2] == mark and board[5] == mark and board[8] == mark) or (board[2] == mark and board[4] == mark and board[6] == mark) or (board[0] == mark and board[4] == mark and board[8] == mark)):
                return True
        else:
                return False
                
def game_on():
        global XorO_P1, XorO_P2, board, player1, player2
        turn = 1
        win = False
        vacancy = True
        while(win==False and vacancy==True):
                if(turn%2==1):
                        inp = input("Player 1 enter marker position : ")
                        input1 = int(inp)
                        if board[input1-1]==" ":
                                board[input1-1] = XorO_P1
                                turn+=1
                                win = win_check(board, inp)
                                print(win)
                                print_board(board)
                                if win==True:
                                        print("Player 1 has won !")
                        else:
                                print("Invalid Position")
                        

                else:
                        inp = input("Player 2 enter marker position : ")
                        input2 = int(inp)
                        if board[input2-1]==" ":
                                board[input2-1] = XorO_P2
                                turn+=1
                                win = win_check(board, inp)
                                print(win)
                                print_board(board)
                                if win==True:
                                        print("Player 2 has won !")
                        else:
                                print("Invalid Position")


                if " " in board:
                        vacancy = True
                else:
                        vacancy = False
        if (vacancy == False and win == True):
                print("Match Draw !")

                        


choose_player()
game_on()
```
